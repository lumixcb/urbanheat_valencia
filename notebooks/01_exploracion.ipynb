```python
# 📦 Librerías
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# 📁 Rutas
csv_path = "../data/raw/temperatura_valencia.csv"
geojson_path = "../data/raw/vegetacion_valencia.geojson"

# 📊 Cargar datos de temperatura
print("⏳ Cargando datos de temperatura...")
df_temp = pd.read_csv(csv_path, sep=';')  # ya separando por ';'
print("✅ Datos de temperatura cargados y procesados:")
display(df_temp.head())

# 🔧 Reemplazar comas por puntos y convertir a float
columnas_float = ["tmed", "tmax", "tmin", "prec", "velmedia", "racha"]

for col in columnas_float:
    df_temp[col] = df_temp[col].str.replace(",", ".", regex=False)
    df_temp[col] = pd.to_numeric(df_temp[col], errors="coerce")

print("\n✅ Conversión de columnas numéricas completada:")
display(df_temp[columnas_float].describe())

# 📌 Verificación
print("✅ Datos de temperatura cargados y procesados:")
display(df_temp.head())
print("\n🔸 Información general:")
print(df_temp.info())

# 🗺️ Cargar datos de vegetación
print("\n⏳ Cargando datos de vegetación...")
gdf_veg = gpd.read_file(geojson_path)

# 📌 Verificación
print("✅ Datos de vegetación cargados:")
display(gdf_veg.head())
print("\n🔸 Información general:")
print(gdf_veg.info())

# 🖼️ Mapa simple de cobertura vegetal
print("\n🗺️ Mostrando mapa de vegetación...")
gdf_veg.plot(figsize=(10, 10), color="green", alpha=0.5)
plt.title("Cobertura vegetal en València")
plt.axis("off")
plt.show()
```

```python
df_temp['fecha'] = pd.to_datetime(df_temp['fecha'])
df_temp['año'] = df_temp['fecha'].dt.year
df_temp['mes'] = df_temp['fecha'].dt.month

df_temp.groupby("año")["tmax"].mean().plot(kind="bar", title="Temperatura máxima media por año", color="salmon")
plt.ylabel("°C")
plt.show()
```

```python
df_temp["fecha"] = pd.to_datetime(df_temp["fecha"])
df_temp.set_index("fecha")["tmed"].plot(figsize=(12, 4), alpha=0.7)
plt.title("Evolución diaria de la temperatura media en València")
plt.ylabel("°C")
plt.xlabel("Fecha")
plt.grid(True)
plt.show()
```

```python
df_temp["mes"] = df_temp["fecha"].dt.month
df_temp.groupby("mes")["tmed"].mean().plot(kind="bar", color="skyblue")
plt.title("Temperatura media por mes (media de 2018–2022)")
plt.xlabel("Mes")
plt.ylabel("°C")
plt.show()
```

```python
df_temp["delta_temp"] = df_temp["tmax"] - df_temp["tmin"]
df_temp["delta_temp"].hist(bins=20, color="goldenrod", edgecolor="black")
plt.title("Diferencia diaria entre temperatura máxima y mínima")
plt.xlabel("Δ Temperatura (°C)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df_temp["año"] = df_temp["fecha"].dt.year
dias_lluvia = df_temp[df_temp["prec"] > 0].groupby("año").size()
dias_lluvia.plot(kind="bar", color="cornflowerblue")
plt.title("Número de días con lluvia por año")
plt.ylabel("Días")
plt.xlabel("Año")
plt.show()
```

```python
df_temp["dir"].hist(bins=36, color="lightseagreen")
plt.title("Distribución de direcciones del viento")
plt.xlabel("Dirección (grados)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Convertir columnas datetime a string para evitar errores en la serialización
gdf_veg_safe = gdf_veg.copy()
for col in gdf_veg_safe.select_dtypes(include=["datetime64[ns, UTC]", "datetime64[ns]"]).columns:
    gdf_veg_safe[col] = gdf_veg_safe[col].astype(str)

# Mapa interactivo
import folium
from folium.plugins import Fullscreen

m = folium.Map(location=[39.47, -0.376], zoom_start=12, tiles="cartodbpositron")
Fullscreen().add_to(m)

# Capa de vegetación
folium.GeoJson(gdf_veg_safe, name="Vegetación", tooltip="elemento").add_to(m)

folium.LayerControl().add_to(m)
m
```
