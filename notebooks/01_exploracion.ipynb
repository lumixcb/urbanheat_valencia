```python
# üì¶ Librer√≠as
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# üìÅ Rutas
csv_path = "../data/raw/temperatura_valencia.csv"
geojson_path = "../data/raw/vegetacion_valencia.geojson"

# üìä Cargar datos de temperatura
print("‚è≥ Cargando datos de temperatura...")
df_temp = pd.read_csv(csv_path, sep=';')  # ya separando por ';'
print("‚úÖ Datos de temperatura cargados y procesados:")
display(df_temp.head())

# üîß Reemplazar comas por puntos y convertir a float
columnas_float = ["tmed", "tmax", "tmin", "prec", "velmedia", "racha"]

for col in columnas_float:
    df_temp[col] = df_temp[col].str.replace(",", ".", regex=False)
    df_temp[col] = pd.to_numeric(df_temp[col], errors="coerce")

print("\n‚úÖ Conversi√≥n de columnas num√©ricas completada:")
display(df_temp[columnas_float].describe())

# üìå Verificaci√≥n
print("‚úÖ Datos de temperatura cargados y procesados:")
display(df_temp.head())
print("\nüî∏ Informaci√≥n general:")
print(df_temp.info())

# üó∫Ô∏è Cargar datos de vegetaci√≥n
print("\n‚è≥ Cargando datos de vegetaci√≥n...")
gdf_veg = gpd.read_file(geojson_path)

# üìå Verificaci√≥n
print("‚úÖ Datos de vegetaci√≥n cargados:")
display(gdf_veg.head())
print("\nüî∏ Informaci√≥n general:")
print(gdf_veg.info())

# üñºÔ∏è Mapa simple de cobertura vegetal
print("\nüó∫Ô∏è Mostrando mapa de vegetaci√≥n...")
gdf_veg.plot(figsize=(10, 10), color="green", alpha=0.5)
plt.title("Cobertura vegetal en Val√®ncia")
plt.axis("off")
plt.show()
```

```python
df_temp['fecha'] = pd.to_datetime(df_temp['fecha'])
df_temp['a√±o'] = df_temp['fecha'].dt.year
df_temp['mes'] = df_temp['fecha'].dt.month

df_temp.groupby("a√±o")["tmax"].mean().plot(kind="bar", title="Temperatura m√°xima media por a√±o", color="salmon")
plt.ylabel("¬∞C")
plt.show()
```

```python
df_temp["fecha"] = pd.to_datetime(df_temp["fecha"])
df_temp.set_index("fecha")["tmed"].plot(figsize=(12, 4), alpha=0.7)
plt.title("Evoluci√≥n diaria de la temperatura media en Val√®ncia")
plt.ylabel("¬∞C")
plt.xlabel("Fecha")
plt.grid(True)
plt.show()
```

```python
df_temp["mes"] = df_temp["fecha"].dt.month
df_temp.groupby("mes")["tmed"].mean().plot(kind="bar", color="skyblue")
plt.title("Temperatura media por mes (media de 2018‚Äì2022)")
plt.xlabel("Mes")
plt.ylabel("¬∞C")
plt.show()
```

```python
df_temp["delta_temp"] = df_temp["tmax"] - df_temp["tmin"]
df_temp["delta_temp"].hist(bins=20, color="goldenrod", edgecolor="black")
plt.title("Diferencia diaria entre temperatura m√°xima y m√≠nima")
plt.xlabel("Œî Temperatura (¬∞C)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df_temp["a√±o"] = df_temp["fecha"].dt.year
dias_lluvia = df_temp[df_temp["prec"] > 0].groupby("a√±o").size()
dias_lluvia.plot(kind="bar", color="cornflowerblue")
plt.title("N√∫mero de d√≠as con lluvia por a√±o")
plt.ylabel("D√≠as")
plt.xlabel("A√±o")
plt.show()
```

```python
df_temp["dir"].hist(bins=36, color="lightseagreen")
plt.title("Distribuci√≥n de direcciones del viento")
plt.xlabel("Direcci√≥n (grados)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Convertir columnas datetime a string para evitar errores en la serializaci√≥n
gdf_veg_safe = gdf_veg.copy()
for col in gdf_veg_safe.select_dtypes(include=["datetime64[ns, UTC]", "datetime64[ns]"]).columns:
    gdf_veg_safe[col] = gdf_veg_safe[col].astype(str)

# Mapa interactivo
import folium
from folium.plugins import Fullscreen

m = folium.Map(location=[39.47, -0.376], zoom_start=12, tiles="cartodbpositron")
Fullscreen().add_to(m)

# Capa de vegetaci√≥n
folium.GeoJson(gdf_veg_safe, name="Vegetaci√≥n", tooltip="elemento").add_to(m)

folium.LayerControl().add_to(m)
m
```
